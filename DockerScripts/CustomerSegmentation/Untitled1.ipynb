```python

%matplotlib inline
import matplotlib.pyplot as plt
v_axis = 1.5
h_line = 2

iYMax = 2.4
iYMin = 1

iXMax = 3.4
iXMin = .5



T_Cutoff = v_axis
V_Cutoff = h_line

fig,ax = plt.subplots(figsize=(15,12))
ax.set_xlim(iXMin,iXMax)
ax.set_ylim(iYMin,iYMax)


ax.axvline(T_Cutoff,color = 'black',linestyle='dashed',lw=2)
ax.axhline(V_Cutoff,color = 'black',linestyle='dashed',lw=2)

ax.fill_between([iXMin, v_axis],iYMin,h_line,alpha=0.3, color='#1F98D0')  # blue
ax.fill_between([v_axis, iXMax], iYMin, h_line, alpha=0.3, color='#F9D307')  # yellow
ax.fill_between([iXMin, v_axis], h_line, iYMax, alpha=0.3, color='#F38D25')  # orange
ax.fill_between([v_axis, iXMax], h_line, iXMax, alpha=0.3, color='#DA383D')  # red

plt.show()
```

```python
df_simei = df[df['street_name'].isin(streets_simei)]
df_simei = df_simei[df_simei['block'].isin(blockno_simei)]
print('simei', len(df_simei))

df_tampines = df[df['street_name'].isin(streets_tampines)]
df_tampines = df_tampines[df_tampines['block'].isin(blockno_tampines)]
print('tampines', len(df_tampines))

df_bedok = df[df['street_name'].isin(streets_bedok)]
df_bedok = df_bedok[df_bedok['block'].isin(blockno_bedok)]
print('bedok', len(df_bedok))

df_tenaga = df[df['street_name'].isin(streets_tenaga)]
df_tenaga = df_tenaga[df_tenaga['block'].isin(blockno_tenaga)]
print('tenaga', len(df_tenaga))
```

```python
df_flats_near_dtl = pd.concat([
    df[df['street_name'].isin(streets_near_dtl)],
    df_simei,
    df_tampines,
    df_bedok,
    df_tenaga
])
print('Dataset length:', len(df_flats_near_dtl))
```

```python
df_flats_near_dtl['month'].unique()
```

```python
df_flats_near_dtl_prices = pd.DataFrame(columns=['Month', 'Price'])
```

```python
months = df_flats_near_dtl['month'].unique()
months.sort()
for i in range(len(months)):
    df_temp = df_flats_near_dtl.loc[df_flats_near_dtl['month'] == months[i]]
    df_flats_near_dtl_prices.loc[i] = [months[i], df_temp['resale_price'].mean()]
    #print(months[i], ':', len(df_flats_near_dtl.loc[df_flats_near_dtl['month'] == months[i]]))

df_flats_near_dtl_prices.info()
df_flats_near_dtl_prices
```

```python
fig, ax = plt.subplots(figsize=(16,10))
fig.autofmt_xdate()
#ax.xaxis.set_ticks(np.arange(min(df_flats_near_dtl_prices['Month']), max(df_flats_near_dtl_prices['Month']), 3))
ax.plot(df_flats_near_dtl_prices['Month'], df_flats_near_dtl_prices['Price'])
ax.set(xlabel='Month', ylabel='Resale price')
```

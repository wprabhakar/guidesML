There are 5 separate cases:

2. CCL East (Mountbatten, Dakota, MacPherson)
3. CCL West (Caldecott, Farrer Road, Holland Village, Buona Vista, Pasir Panjang, Labrador Park, Telok Blangah)

### Data Cleaning

As there are now 6 DataFrames, we include a common `clean_data()` function.

The following cleaning methods will be employed:
- Drop `town`, `block`, `street_name`, `storey_range`, `floor_area_sqm`, `flat_model`, `lease_commence_date`
- Only keep the rows where `room_type` is
    - 3 ROOM
    - 4 ROOM
    - 5 ROOM
- Replace `month` column with `year` column (to minimise the effects of short-term fluctuations)

```python
def clean_data(df):
    df.drop(['town', 'block', 'street_name', 'storey_range', 'floor_area_sqm', 'flat_model', 'lease_commence_date'], 
            axis=1, inplace=True)
    
    #df = df.loc[(df.flat_type == '3 ROOM') | (df.flat_type == '4 ROOM') | (df.flat_type == '5 ROOM')]
    
    # add a 'year' column
    month_split = df['month'].str.split('-', expand=True)
    df['year'] = month_split[0]
    # add a 'quarter' column
    df['quarter'] = pd.to_datetime(df['month'])
    df['quarter'] = pd.PeriodIndex(df['quarter'], freq='Q')
    df.quarter = df.quarter.astype(str)
    # drop the 'month' column
    df.drop('month', axis=1, inplace=True)
    
    df.sort_values(['year'], inplace=True)
    return df   
```

```python
# all 6 dataframes to call the clean_data function

df = clean_data(df)
df_ccl_east = clean_data(df_ccl_east)
df_ccl_north = clean_data(df_ccl_north)
df_ccl_west = clean_data(df_ccl_west)
df_dtl_east = clean_data(df_dtl_east)
df_dtl_west = clean_data(df_dtl_west)
```

### Data Exploration

We wish to plot the price trends for each scenario and observe if there is a discernible difference in the price movement compared to the national average in the period after the opening of a nearby train station.

The different periods where a train station was opened for the 5 cases are:
1. CCL North - May 2009
2. CCL East - Apr 2010
3. CCL West - Oct 2011
4. DTL West - Dec 2015
5. DTL East - Oct 2017

First, we create a secondary associated DataFrame for all the cases. This DataFrame has 2 columns:
1. Year
2. Average Resale Price

```python
def create_df_prices(df):
    df_prices_yr = pd.DataFrame(columns=['year', 'avg_resale_price', 'percentage_change_price'])
    df_prices_qtr = pd.DataFrame(columns=['quarter', 'avg_resale_price', 'percentage_change_price'])
    
    years = df['year'].unique()
    for i in range(len(years)):
        df_temp = df.loc[df['year'] == years[i]]
        df_prices_yr.loc[i] = [years[i], df_temp['resale_price'].mean(), '']
    
    quarters = df['quarter'].unique()
    for i in range(len(quarters)):
        df_temp = df.loc[df['quarter'] == quarters[i]]
        df_prices_qtr.loc[i] = [quarters[i], df_temp['resale_price'].mean(), '']
    
    df_prices_yr['percentage_change_price'] = df_prices_yr['avg_resale_price'].pct_change()
    df_prices_qtr['percentage_change_price'] = df_prices_qtr['avg_resale_price'].pct_change()
    
    return df_prices_yr, df_prices_qtr
```

```python
df_prices_yr, df_prices_qtr = create_df_prices(df)
df_ccl_east_prices_yr, df_ccl_east_prices_qtr = create_df_prices(df_ccl_east)
df_ccl_north_prices_yr, df_ccl_north_prices_qtr = create_df_prices(df_ccl_north)
df_ccl_west_prices_yr, df_ccl_west_prices_qtr = create_df_prices(df_ccl_west)
df_dtl_east_prices_yr, df_dtl_east_prices_qtr = create_df_prices(df_dtl_east)
df_dtl_west_prices_yr, df_dtl_west_prices_qtr = create_df_prices(df_dtl_west)
```

Then, we plot the graphs for each case compared to the national average.

```python
def plot_price_change_with_national_avg_yr(df, df_national, case):
    # for plotting of vertical line
    train_opening_year = {
        'CCL North': 9,
        'CCL East': 10,
        'CCL West': 11,
        'DTL West': 16,
        'DTL East': 17
    }
    fig, ax = plt.subplots(figsize=(14, 8), nrows=2, ncols=1)
    
    ax[0].plot(df_national['year'], df_national['avg_resale_price'], label='National', color='r')
    ax[0].plot(df['year'], df['avg_resale_price'], label=case, color='g')
    ax[0].axvline(x=train_opening_year[case])
    ax[0].set(ylabel='Average resale price')
    ax[0].legend()
    
    ax[1].plot(df_national['year'], df_national['percentage_change_price'], label='National', color='r')
    ax[1].plot(df['year'], df['percentage_change_price'], label=case, color='g')
    ax[1].axvline(x=train_opening_year[case])
    ax[1].axhline(y=0, color='black', linewidth=1, linestyle='--')
    ax[1].set(ylabel='Percentage change in average resale price')
    ax[1].legend()
    
def plot_price_change_with_national_avg_qtr(df, df_national, case):
    # for plotting of vertical line
    train_opening_year = {
        'CCL North': 9,
        'CCL East': 10,
        'CCL West': 11,
        'DTL West': 16,
        'DTL East': 17
    }
    fig, ax = plt.subplots(figsize=(14, 8), nrows=2, ncols=1)
    
    ax[0].plot(df_national['quarter'], df_national['avg_resale_price'], label='National', color='r')
    ax[0].plot(df['quarter'], df['avg_resale_price'], label=case, color='g')
    ax[0].axvline(x=train_opening_year[case])
    ax[0].set(ylabel='Average resale price')
    ax[0].legend()
    
    ax[1].plot(df_national['quarter'], df_national['percentage_change_price'], label='National', color='r')
    ax[1].plot(df['quarter'], df['percentage_change_price'], label=case, color='g')
    ax[1].axvline(x=train_opening_year[case])
    ax[1].axhline(y=0, color='black', linewidth=1, linestyle='--')
    ax[1].set(ylabel='Percentage change in average resale price')
    ax[1].legend()
```
